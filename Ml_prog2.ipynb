Develop a program to Compute the correlation matrix to
understand the relationships between pairs of features. Visualize
the correlation matrix using a heatmap to know which variables
have strong positive/negative correlations. Create a pair plot to
visualize pairwise relationships between features. Use California
Housing dataset.

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cal_data = fetch_california_housing(as_frame=True)
data=cal_data.frame
correlation_matrix=data.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("correlation matrix")
plt.show()
```

```python
sns.pairplot(data, diag_kind='hist', plot_kws={'alpha':0.5})
plt.suptitle("Pair plot of CHD")
plt.show()
```
