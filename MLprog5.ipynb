Develop a program to implement k-Nearest Neighbour algorithm
to classify the randomly generated 100
values of x in the range of [0,1]. Perform the following based on
dataset generated.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
np.random.seed(42)
```

```python
x_values=np.random.rand(100,1)
y_labels = np.array(['Class1' if x<=0.5 else 'Class2' for x in x_values.flatten()])
X_train = x_values[:50]
y_train = y_labels[:50]
X_test = x_values[50:]
y_test = y_labels[50:]
```

```python
k_values = [1, 2, 3, 4, 5, 20, 30]
```

```python

for i,k in enumerate(k_values,1):
    plt.figure(figsize=(12,8))
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    plt.subplot(3,3,i)
    plt.title(f"KNN classifier plot for k = {k}")
    plt.scatter(X_test, y_test, color='blue', label="True")
    plt.scatter(X_test, y_pred, color='red', marker='x', label="predicted")
    plt.legend(loc='best')
    plt.grid(True)
    plt.plot()
    plt.show()
```
