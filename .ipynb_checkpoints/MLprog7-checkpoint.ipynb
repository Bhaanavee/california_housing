Develop a program to demonstrate the working of Linear
Regression and Polynomial Regression. Use
Boston Housing Dataset for Linear Regression and Auto MPG
Dataset (for vehicle fuel efficiency prediction)
for Polynomial Regression.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
```

```python
# ==== LINEAR REGRESSION: Boston Housing ====
boston = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")
X = boston[['rm']]  # Feature: average rooms
y = boston['medv']  # Target: house price
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# Plot Boston Linear
plt.subplot(1, 2, 1)
plt.scatter(X_test, y_test, label='Actual')
plt.plot(X_test, y_pred, color='red', label='Predicted')
plt.title("Boston Linear")
plt.xlabel("Rooms")
plt.ylabel("Price")
plt.legend()
```

```python
# ==== POLYNOMIAL REGRESSION: Auto MPG ====
auto = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv")
auto = auto.dropna()
X = auto[['horsepower']]
y = auto['mpg']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Make it polynomial
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
model = LinearRegression()
model.fit(X_train_poly, y_train)
y_pred = model.predict(X_test_poly)

# Plot Auto MPG Polynomial
plt.subplot(1, 2, 2)
plt.scatter(X_test, y_test, label='Actual')
plt.scatter(X_test, y_pred, color='red', label='Predicted', alpha=0.6)
plt.title("Auto MPG Polynomial")
plt.xlabel("Horsepower")
plt.ylabel("MPG")
plt.legend()

plt.tight_layout()
plt.show()
```
